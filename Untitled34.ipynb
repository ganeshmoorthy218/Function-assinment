```python
#Q11.Identify and remove outliers in the km_driven column using the IQR method.
import pandas as pd

# Load dataset
df = pd.read_csv("BIKE DETAILS.csv")
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['km_driven'] < lower_bound) | (df['km_driven'] > upper_bound)]
print(f"Number of outliers detected: {len(outliers)}")

# Remove outliers
df_no_outliers = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]

print(f"Data shape before removing outliers: {df.shape}")
print(f"Data shape after removing outliers: {df_no_outliers.shape}")
```

```python
 #Q14.Which bike names are priced significantly above the average price for their manufacturing year?
 import pandas as pd

# Load your dataset
df = pd.read_csv("BIKE DETAILS.csv")

# Step 1: Calculate average selling price per manufacturing year
avg_price_per_year = df.groupby('year')['selling_price'].mean().reset_index()
avg_price_per_year.rename(columns={'selling_price': 'avg_price_year'}, inplace=True)

# Step 2: Merge average price back to original dataframe
df = df.merge(avg_price_per_year, on='year', how='left')

# Step 3: Calculate price difference (absolute and percentage)
df['price_diff'] = df['selling_price'] - df['avg_price_year']
df['price_diff_pct'] = df['price_diff'] / df['avg_price_year'] * 100

# Step 4: Define threshold (e.g., 20% above average)
threshold = 20

# Step 5: Filter bikes priced significantly above average
above_avg_bikes = df[df['price_diff_pct'] > threshold]

# Show bike names, year, selling price, average price, and difference
result = above_avg_bikes[['name', 'year', 'selling_price', 'avg_price_year', 'price_diff_pct']]

print(result.sort_values(by='price_diff_pct', ascending=False))
```

```python
#Q1.What is the average rating for each product category?
import pandas as pd

# Load dataset
df = pd.read_csv("amazon.csv")

# Convert 'rating' column to numeric, coercing errors to NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Calculate average rating per category, dropping rows with NaN ratings
# We dropna here to ensure the mean is calculated only on valid numeric ratings
avg_rating_per_category = df.dropna(subset=['rating']).groupby('category')['rating'].mean().reset_index()

# Rename columns for clarity
avg_rating_per_category.columns = ['Category', 'Average Rating']

# Sort by average rating descending
avg_rating_per_category = avg_rating_per_category.sort_values(by='Average Rating', ascending=False)

print(avg_rating_per_category)
```

```python
#Q4.How does the average discount percentage vary across categories?
import pandas as pd

# Load the dataset
df = pd.read_csv("amazon.csv")

# --- Start of Suggested Changes ---

# Clean the 'discount_percentage' column
# Remove the '%' sign and convert to numeric (float)
# Use errors='coerce' to turn any unparseable values into NaN
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '', regex=False)
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# --- End of Suggested Changes ---


# Calculate average discount percentage per category
# We can now safely calculate the mean because the column is numeric
# .mean() will automatically handle NaN values by default (skipping them)
avg_discount_per_category = df.groupby('category')['discount_percentage'].mean().reset_index()

# Rename columns for clarity
avg_discount_per_category.columns = ['Category', 'Average Discount Percentage']

# Sort by average discount descending (optional)
avg_discount_per_category = avg_discount_per_category.sort_values(by='Average Discount Percentage', ascending=False)

print(avg_discount_per_category)
```

```python
#Q5.What are the most popular product names?
import pandas as pd

# Load dataset
df = pd.read_csv("amazon.csv")  # Update with your file path

# Convert 'rating_count' to numeric, coercing errors to NaN
# This handles cases where rating_count might not be purely numeric
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Fill NaN values in 'rating_count' with 0 before summing
# This ensures sum() works correctly and we don't lose products with missing counts
df['rating_count'] = df['rating_count'].fillna(0)

# Group by product_name and sum rating_count
popularity = df.groupby('product_name')['rating_count'].sum().reset_index()

# Sort products by rating_count descending
# Now that rating_count is numeric, sort_values should work correctly
top_popular_products = popularity.sort_values(by='rating_count', ascending=False)

# Show top 10 most popular products
print(top_popular_products.head(10))
```

```python
#Q6.What are the most popular product keywords?
import pandas as pd
from collections import Counter
import re

# Load dataset
df = pd.read_csv("amazon.csv")  # Update path if needed

# Function to clean and split product names into keywords
def extract_keywords(text):
    # Lowercase, remove non-alphanumeric characters, split by whitespace
    words = re.findall(r'\b\w+\b', str(text).lower())
    # You can add stopword removal here if needed
    return words

# --- Start of Suggested Changes ---

# Convert 'rating_count' to string first to handle potential non-string types
df['rating_count'] = df['rating_count'].astype(str)

# Remove commas from the 'rating_count' column
df['rating_count'] = df['rating_count'].str.replace(',', '', regex=False)

# Convert 'rating_count' to numeric (integer), coercing errors to NaN
# Use errors='coerce' to turn any unparseable values into NaN
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Fill NaN values in 'rating_count' with 0
# This ensures sum() works correctly and we don't lose products with missing counts
df['rating_count'] = df['rating_count'].fillna(0)

# --- End of Suggested Changes ---


# Create a list of (keyword, rating_count) tuples
keywords_with_counts = []

for _, row in df.iterrows():
    keywords = extract_keywords(row['product_name'])
    # Now rating_count is guaranteed to be numeric (or 0)
    rating_count = row['rating_count']
    for kw in keywords:
        keywords_with_counts.append((kw, rating_count))

# Aggregate rating_counts by keyword
keyword_counter = Counter()
for kw, count in keywords_with_counts:
    # Now 'count' is a number, so += operation is valid
    keyword_counter[kw] += count

# Get top 20 keywords by weighted popularity
top_keywords = keyword_counter.most_common(20)

print("Top 20 Popular Product Keywords (weighted by rating_count):")
for kw, cnt in top_keywords:
    print(f"{kw}: {cnt}")
```

```python
#Q8.What is the correlation between discounted_price and rating?
import pandas as pd
import numpy as np # Import numpy for np.nan

# Load dataset
# Assuming 'amazon.csv' is in the correct path
try:
    df = pd.read_csv("amazon.csv")
except FileNotFoundError:
    print("Error: amazon.csv not found. Please check the file path.")
    # Exit or handle the error appropriately if the file is not found
    exit()


# Clean the 'discounted_price' column
# Convert to string first to handle potential non-string types before replacement
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)

# Convert 'discounted_price' to numeric, coercing errors to NaN
# Use errors='coerce' to turn any unparseable values into NaN
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

# Ensure 'rating' is numeric as the correlation involves both columns
# Convert to string first to handle potential non-string types before replacement
df['rating'] = df['rating'].astype(str)

# Remove any non-numeric characters that might still be in 'rating' (e.g., trailing spaces, non-decimal points)
# A more robust approach might be needed if 'rating' contains non-standard representations
# For now, we'll just strip whitespace and hope pd.to_numeric handles decimals
df['rating'] = df['rating'].str.strip()

# Convert 'rating' to numeric, coercing errors to NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')


# Drop rows with NaN values in either column before calculating correlation
# Correlation requires non-null values
df_cleaned_corr = df.dropna(subset=['discounted_price', 'rating'])

# Check if there are enough data points after cleaning
if len(df_cleaned_corr) < 2:
    print("Not enough valid data points to calculate correlation after cleaning.")
    correlation = np.nan # Assign NaN if correlation cannot be calculated
else:
    # Calculate correlation between discounted_price and rating
    correlation = df_cleaned_corr['discounted_price'].corr(df_cleaned_corr['rating'])


print(f"Correlation between discounted_price and rating: {correlation:.4f}")
```

```python
#Q10. Identify any potential areas for improvement or optimization based on the data analysis.
import pandas as pd
import re
from collections import Counter
import numpy as np # Import numpy for np.nan

# Load dataset
df = pd.read_csv("amazon.csv")

# --- Cleaning steps added ---
# Clean the 'discount_percentage' column
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '', regex=False)
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# Clean the 'rating' column
df['rating'] = df['rating'].astype(str).str.strip() # Strip whitespace
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Clean the 'review_id' column (assuming it might have non-numeric issues for count aggregation)
# Convert to string first to handle potential non-string types before replacement
df['review_id'] = df['review_id'].astype(str)
# Remove commas from the 'review_id' column
df['review_id'] = df['review_id'].str.replace(',', '', regex=False)
# Convert 'review_id' to numeric (integer), coercing errors to NaN
df['review_id'] = pd.to_numeric(df['review_id'], errors='coerce')
# Fill NaN values in 'review_id' with 0
df['review_id'] = df['review_id'].fillna(0)

# Clean the 'discounted_price' column
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

# --- End of Cleaning steps ---


# 1. Categories with lowest discounts
# Ensure discount_percentage is numeric and handle potential NaNs before calculating mean
low_discount = df.dropna(subset=['discount_percentage']).groupby('category')['discount_percentage'].mean().nsmallest(3)

# 2. Top rated categories
# Ensure rating is numeric and handle potential NaNs before calculating mean
top_rated = df.dropna(subset=['rating']).groupby('category')['rating'].mean().nlargest(3)

# 3. Products with many reviews but low ratings
# Ensure rating and review_id are numeric and handle potential NaNs
prod_stats = df.dropna(subset=['rating', 'review_id']).groupby('product_name').agg(
    {'rating':'mean', 'review_id':'sum'} # Use sum for review_id count aggregation
)
# Filter based on quantiles of the summed review_id
problem_products = prod_stats[(prod_stats['review_id'] >= prod_stats['review_id'].quantile(0.75)) & (prod_stats['rating'] < 3.5)]

# 4. Correlation between discounted price and rating
# Ensure both columns are numeric and handle potential NaNs
df_cleaned_corr = df.dropna(subset=['discounted_price', 'rating'])
# Check if there are enough data points after cleaning
if len(df_cleaned_corr) < 2:
    corr = np.nan # Assign NaN if correlation cannot be calculated
else:
    corr = df_cleaned_corr['discounted_price'].corr(df_cleaned_corr['rating'])


# 5. Top keywords in product names
keywords = Counter()
for name in df['product_name'].dropna():
    keywords.update(re.findall(r'\b\w+\b', name.lower()))
top_keywords = keywords.most_common(10)

print("Low discount categories:\n", low_discount)
print("\nTop rated categories:\n", top_rated)
print("\nProblematic products:\n", problem_products)
print(f"\nPrice-Rating correlation: {corr:.3f}")
print("\nTop keywords:", top_keywords)
```

```python
#Q7.For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist?
import pandas as pd

# Load dataset
df = pd.read_csv("spotify.csv")

# Step 1: Compute average popularity per artist
artist_avg_popularity = df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False)

# Step 2: Get top 5 most popular artists
top_5_artists = artist_avg_popularity.head(5).index

# Step 3: For each of these artists, find their most popular track
most_popular_tracks = []

for artist in top_5_artists:
    artist_tracks = df[df['Artist'] == artist]
    top_track = artist_tracks.loc[artist_tracks['Popularity'].idxmax()]
    most_popular_tracks.append((artist, top_track['Track Name'], top_track['Popularity']))

# Step 4: Display the results
print("Most Popular Track for Each of the Top 5 Artists:\n")
for artist, track, popularity in most_popular_tracks:
    print(f"Artist: {artist}\n  Track: {track}\n  Popularity: {popularity}\n")
```

```python
#Q9.Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("spotify.csv")

# Get top 10 artists by track count
top_artists = df['Artist'].value_counts().head(10).index

# Filter dataset for those artists
df_top = df[df['Artist'].isin(top_artists)]

# Plot: Box plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_top, x='Artist', y='Duration (ms)', palette='Set2')
plt.title("Track Duration Distribution Across Top 10 Artists (Box Plot)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Optional: Violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(data=df_top, x='Artist', y='Duration (ms)', palette='Set3')
plt.title("Track Duration Distribution Across Top 10 Artists (Violin Plot)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Q10.How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("spotify.csv")

# Select top 10 artists by number of tracks
top_artists = df['Artist'].value_counts().head(10).index
df_top = df[df['Artist'].isin(top_artists)]

# Violin Plot: Popularity distribution per artist
plt.figure(figsize=(12, 6))
sns.violinplot(data=df_top, x='Artist', y='Popularity', palette='pastel')
plt.title("Popularity Distribution Across Top 10 Artists (Violin Plot)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Optional: Swarm Plot for detailed points
plt.figure(figsize=(12, 6))
sns.swarmplot(data=df_top, x='Artist', y='Popularity', color='black', alpha=0.6, size=3)
plt.title("Popularity Distribution Across Top 10 Artists (Swarm Plot)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
