# **#Theoretical**

# **#Q1.What is a Decision Tree, and how does it work?**

A Decision Tree is a supervised machine learning algorithm used for both classification and regression tasks. It works by recursively partitioning the data into subsets based on the most significant features, creating a tree-like model of decisions and their possible consequences. Each internal node represents a "test" on an attribute, each branch represents the outcome of the test, and each leaf node represents a class label (in classification) or a predicted value (in regression).

# **#Q2.What are impurity measures in Decision Trees?**

Impurity measures are metrics used in Decision Trees to quantify the "mixed-up" nature of a set of samples. The goal of a Decision Tree is to split the data at each node in a way that minimizes the impurity of the resulting child nodes. Common impurity measures include Gini Impurity and Entropy.

# **#Q5.What is Information Gain, and how is it used in Decision Trees?**

Information Gain is the reduction in entropy (or impurity) achieved by splitting a dataset on a particular attribute. It measures how much information a feature provides about the class. In Decision Trees, the algorithm selects the attribute that yields the highest Information Gain at each step to make the split, aiming to create the purest possible child nodes.

# **#Q6.What is the difference between Gini Impurity and Entropy?**

Both Gini Impurity and Entropy are measures of impurity used to evaluate splits in Decision Trees.

Gini Impurity measures the probability of incorrectly classifying a randomly chosen element from the set if it were randomly labeled according to the distribution of labels in the subset. It favors larger partitions.

Entropy measures the average amount of information needed to identify the class of a randomly chosen element from the set. It tends to produce more balanced trees.
In practice, both often yield similar results, though Gini Impurity is computationally slightly faster as it doesn't involve logarithms.

# **#Q7.What is the mathematical explanation behind Decision Trees?**

The mathematical explanation behind Decision Trees revolves around finding optimal splits that maximize information gain or minimize impurity. At each node, the algorithm evaluates all possible splits for all features. For a given feature and a potential split point (for continuous features) or set of categories (for categorical features), it calculates the impurity of the resulting child nodes. The split that leads to the greatest reduction in impurity (or highest information gain) is chosen. This process is recursively applied until a stopping criterion is met (e.g., maximum depth, minimum samples per leaf, or no further impurity reduction).

# **#Q8.What is Pre-Pruning in Decision Trees?**

Pre-pruning (also known as early stopping) involves stopping the growth of the Decision Tree before it has fully grown. This is done by setting certain stopping criteria during the tree construction phase, such as:

Maximum depth of the tree

Minimum number of samples required to split a node

Minimum number of samples required in a leaf node

Minimum impurity decrease required for a split

# **#Q9.What is Post-Pruning in Decision Trees?**

Post-pruning involves growing the Decision Tree to its full potential (or a very large size) and then, after the tree is built, removing branches that do not contribute significantly to the predictive accuracy. This is often done by evaluating the performance on a separate validation set and pruning branches that lead to an increase in error on that set. Techniques include reduced error pruning and cost-complexity pruning.

# **#Q10.What is the difference between Pre-Pruning and Post-Pruning?**

Pre-Pruning: Stops tree growth during construction. It's a "top-down" approach. It can be faster as it avoids building an overly complex tree. However, it might stop too early and miss potentially good splits.

Post-Pruning: Prunes the tree after it has been fully grown. It's a "bottom-up" approach. It generally leads to more optimal trees as it considers the entire tree structure but is computationally more expensive.

# **#Q11.What is a Decision Tree Regressor?**

A Decision Tree Regressor is a Decision Tree algorithm adapted for regression tasks (predicting continuous numerical values). Instead of predicting a class label at each leaf node, it predicts a numerical value. For a given leaf node, the predicted value is typically the average (or median) of the target values of all the training samples that fall into that leaf. The splitting criteria for regression trees are usually based on minimizing measures like Mean Squared Error (MSE) or Mean Absolute Error (MAE) instead of impurity.

# **#Q12.What are the advantages and disadvantages of Decision Trees?**

**Advantages:**

Easy to understand and interpret (visualizable).

Can handle both numerical and categorical data.

Requires little data preparation (no normalization needed).

Can perform well even with missing values (if handled properly).

Non-linear relationships between features and target are handled well.

**Disadvantages:**

Prone to overfitting, especially with complex trees.

Can be unstable, as small variations in data can lead to very different tree structures.

Decision boundaries are orthogonal (axis-parallel), which might not be optimal for all data distributions.

Can be biased if some classes dominate.

# **#Q13.How does a Decision Tree handle missing values?**
Decision Trees can handle missing values in several ways:

Ignoring Samples: Samples with missing values for a particular attribute are simply ignored when splitting on that attribute.

Imputation: Missing values can be imputed using various strategies (mean, median, mode, or more sophisticated methods) before building the tree.

Surrogate Splits: Some algorithms (like CART) can use "surrogate splits." If a primary split attribute is missing, a secondary (surrogate) attribute is used that is highly correlated with the primary attribute.

Probabilistic Assignment: A sample with a missing value for a splitting attribute can be sent down all branches, with its weight distributed proportionally to the number of training samples going down each branch

# **#Q14.How does a Decision Tree handle categorical features?**

Decision Trees naturally handle categorical features.

Binary Splits: For a categorical feature with k categories, the tree can create 2

Multi-way Splits: Alternatively, for nominal categorical features, a multi-way split can be created where each branch corresponds to one category.

Ordinal Categories: For ordinal categorical features, they can be treated as numerical and split based on their order. The algorithm evaluates different ways to group these categories to find the split that maximizes information gain or minimizes impurity.

# **#Q15.What are some real-world applications of Decision Trees?**

Decision Trees are widely used in various real-world scenarios due to their interpretability and versatility:

Medical Diagnosis: Diagnosing diseases based on patient symptoms and medical test results.

Credit Risk Assessment: Evaluating the creditworthiness of loan applicants.

Customer Churn Prediction: Identifying customers who are likely to discontinue a service.

Fraud Detection: Detecting fraudulent transactions in banking or insurance.

Marketing Campaign Optimization: Targeting specific customer segments for marketing campaigns.

Fault Diagnosis in Engineering: Identifying the root cause of system failures.

Drug Discovery: Predicting the efficacy of new drug compounds.

```python

#Q16.Write a Python program to train a Decision Tree Classifier on the Iris dataset and print the model accuracy.

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = dt_classifier.predict(X_test)

# Calculate and print the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```

```python
#Q27.Write a Python program to train a Decision Tree Classifier, apply Cost Complexity Pruning (CCP), and visualize its effect on accuracy.

import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
import numpy as np

# 1. Generate a synthetic dataset for demonstration
X, y = make_classification(
    n_samples=1000,
    n_features=4,
    n_informative=2,
    n_redundant=0,
    n_repeated=0,
    n_classes=2,
    n_clusters_per_class=1,
    random_state=42
)

# 2. Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")

# 3. Calculate the cost complexity pruning path
path = DecisionTreeClassifier(random_state=42).cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas # Corrected line
impurities = path.impurities # Corrected line

# Filter out redundant alphas (where impurity doesn't change significantly)
# This helps in getting a clearer plot and avoids training too many identical trees.
ccp_alphas = np.unique(ccp_alphas) # Ensure unique alphas
ccp_alphas = ccp_alphas[ccp_alphas >= 0] # Ensure non-negative alphas

print(f"\nNumber of effective ccp_alphas found: {len(ccp_alphas)}")
print(f"First 5 ccp_alphas: {ccp_alphas[:5]}")
print(f"Last 5 ccp_alphas: {ccp_alphas[-5:]}")

# 4. Train a series of Decision Trees for each ccp_alpha
clfs = []
train_accuracies = []
test_accuracies = []

print("\nTraining Decision Trees with different ccp_alphas...")
for ccp_alpha in ccp_alphas:
    # Initialize a Decision Tree Classifier with the current ccp_alpha
    clf = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
    # Train the classifier on the training data
    clf.fit(X_train, y_train)
    clfs.append(clf)

    # Calculate and store training accuracy
    y_train_pred = clf.predict(X_train)
    train_acc = accuracy_score(y_train, y_train_pred)
    train_accuracies.append(train_acc)

    # Calculate and store testing accuracy
    y_test_pred = clf.predict(X_test)
    test_acc = accuracy_score(y_test, y_test_pred)
    test_accuracies.append(test_acc)

    # Optional: Print progress for larger datasets
    # if len(clfs) % 50 == 0 or ccp_alpha == ccp_alphas[-1]:
    #     print(f"  Trained tree with ccp_alpha={ccp_alpha:.4f}, Train Acc: {train_acc:.4f}, Test Acc: {test_acc:.4f}")

print("\nFinished training all trees.")

# 5. Visualize the effect of ccp_alpha on accuracy
plt.figure(figsize=(12, 6))
plt.plot(ccp_alphas, train_accuracies, marker='o', drawstyle="steps-post", label="Training Accuracy")
plt.plot(ccp_alphas, test_accuracies, marker='o', drawstyle="steps-post", label="Testing Accuracy")

plt.xlabel("ccp_alpha")
plt.ylabel("Accuracy")
plt.title("Accuracy vs. ccp_alpha for Decision Tree Pruning")
plt.legend()
plt.grid(True)
plt.show()

optimal_alpha_index = np.argmax(test_accuracies)
optimal_ccp_alpha = ccp_alphas[optimal_alpha_index]
max_test_accuracy = test_accuracies[optimal_alpha_index]

print(f"\nOptimal ccp_alpha based on testing accuracy: {optimal_ccp_alpha:.4f}")
print(f"Maximum testing accuracy achieved: {max_test_accuracy:.4f}")
```

```python
#Q28.Write a Python program to train a Decision Tree Classifier and evaluate its performance using Precision,Recall, and F1-Score.

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = dt_classifier.predict(X_test)

# Print the classification report
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```

```python
#Q29. Write a Python program to train a Decision Tree Classifier and visualize the confusion matrix using seaborn.

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target
class_names = iris.target_names

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = dt_classifier.predict(X_test)

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Print the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")

# Visualize the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
#Q30.Write a Python program to train a Decision Tree Classifier and use GridSearchCV to find the optimal values for max_depth and min_samples_split.

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the parameter grid for GridSearchCV
param_grid = {
    'max_depth': [None, 3, 5, 10],  # Test different max_depth values, including no limit
    'min_samples_split': [2, 5, 10]  # Test different min_samples_split values
}

# Initialize the Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=dt_classifier, param_grid=param_grid, cv=5, scoring='accuracy')

# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Print the best parameters found
print("Best parameters found by GridSearchCV:")
print(grid_search.best_params_)

# Print the best accuracy score achieved
print("\nBest accuracy score achieved by GridSearchCV:")
print(f"{grid_search.best_score_:.2f}")

# Optionally, evaluate the best estimator on the test set
best_dt_classifier = grid_search.best_estimator_
test_accuracy = best_dt_classifier.score(X_test, y_test)
print(f"\nAccuracy of the best model on the test set: {test_accuracy:.2f}")
```
